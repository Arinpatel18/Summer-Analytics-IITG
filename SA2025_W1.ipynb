### 5. Which year had the highest average mpg?

```python
dict = {}

for idx, row in df.iterrows():
    year = row['model_year']
    mpg = row['mpg']
    dict.setdefault(year, []).append(mpg)
res = 0
maxi = -1
for key in dict:
    sum = 0
    size = 0
    for val in dict[key]:
        sum += val
        size += 1
    avg = sum / size

    if avg>maxi:
        maxi=avg
        res=key
print(res)
```

### We define a “consistent” car model as one that was produced over multiple years and had very low variation in mpg across those years (standard deviation < 1.0).

Tasks:

Identify car names that appear in more than one model_year.

For each such name, compute the standard deviation of mpg across years.

Return the car(s) with the lowest variation in mpg, among those with at least 2 appearances and std(mpg) < 1.0.

Report the model name(s), number of appearances, and the average mpg.

Bonus:
Sort the result by number of appearances (descending), then mpg (descending).

```python
import pandas as pd

# Assuming your DataFrame is named 'df'
# Columns expected: 'car_name', 'model_year', 'mpg'

# Step 1: Group by car_name and count unique model_years
name_years = df.groupby('name')['model_year'].nunique()

# Filter names that appear in more than one model_year
consistent_names = name_years[name_years > 1].index

# Step 2: Filter DataFrame to include only those consistent_names
df_filtered = df[df['name'].isin(consistent_names)]

# Step 3: Group by name and compute standard deviation and count
summary = (
    df_filtered.groupby('name')
    .agg(
        mpg_std=('mpg', 'std'),
        appearances=('model_year', 'nunique'),
        avg_mpg=('mpg', 'mean')
    )
    .reset_index()
)

# Step 4: Filter std(mpg) < 1.0
summary = summary[summary['mpg_std'] < 1.0]

# Step 5: Find minimum mpg_std among those cars
min_std = summary['mpg_std'].min()
result = summary[summary['mpg_std'] == min_std]

# Step 6: Bonus — sort by appearances (descending) then avg_mpg (descending)
result = result.sort_values(by=['appearances', 'avg_mpg'], ascending=[False, False])

# Print the result
print(result[['name', 'appearances', 'avg_mpg']])
```
